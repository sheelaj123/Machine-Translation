BERT model has its way of generating the tokens. It supports a piece models to tokenize the text. It is also possible to use transformer API’t to tokenize.

The below code demonstrates tokenization using both the method in the BERT model

#BERT configuration
It is used to instantiate a BERT model based on the specified arguments.
It is a way to define the model architecture.
Instantiating a configuration with the default results in BERT bert-base-uncased architecture.

```python
from transformers import BertModel, BertConfig
# Initializing a BERT bert-base-uncased style configuration
configuration = BertConfig()
# Initializing a model from the bert-base-uncased style configuration
model = BertModel(configuration)
# Accessing the model configuration
configuration = model.config
```

#BERT tokenizer
#Method 1
BertForPreTraining :

Bert Model with two heads on top as done during the pretraining: a masked language modelling head and a next sentence prediction (classification) head.



```python
from transformers import BertTokenizer, BertModel
import torch
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
inputs = tokenizer("Machine Translation using BERT", return_tensors="pt")
outputs = model(**inputs)
last_hidden_states = outputs.last_hidden_state
```

#Method 2


```python
text = "BERT is the state-of the art technologywhich uses Transformers"
marker = "[CLS] " + text + " [SEP]"
# Tokenize our sentence with the BERT tokenizer.
tokens = tokenizer.tokenize(marker)
# Map the token strings to their vocabulary indeces.
index = tokenizer.convert_tokens_to_ids(tokens)
# Print  the tokens.
print('-------------------')
print('Tokens generated from BERT')
print('-------------------')
print (tokens)
# Display the words with their indeces.
print('------------------------')
print('TOKEN           INDEX')
print('------------------------')
for tup in zip(tokens, index):
    print('{:<12} {:>6,}'.format(tup[0], tup[1]))
```
